```python
import pandas as pd
from pandas.io.formats.excel import ExcelFormatter

from selenium import webdriver
from selenium.webdriver.common.by import By
import time

pesquisa = input('Digite sua busca')

navegador = webdriver.Chrome()
navegador.get("https://www.google.com")

navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(pesquisa)
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[3]/center/input[1]').click()

time.sleep(15)
print('tudo pronto')

#conteudo = navegador.find_element(By.XPATH, '//*[@id="rcnt"]')

lista_empresa = []
lista_telefone = []
resultado = []

#print(conteudo.text) VwiC3b yXK7lf MUxGbd yDYNvb lyLwlc lEBKkf

for numero in range(15):
    print(numero)

    titulo = navegador.find_elements(By.CLASS_NAME, 'LC20lb')
    description = navegador.find_elements(By.CLASS_NAME, 'VwiC3b')
    

    for titulos in titulo:
        string_title = titulos.text
        fatia_titulo = string_title.split('•')
        lista_empresa.append(fatia_titulo[0])
        print(fatia_titulo[0])
    
    print(100 * "#")
    
    for n,descriptions in enumerate(description):
        #print(n,descriptions.text)
        string_description = descriptions.text.lower()
        fatia = string_description.split(' ')
        
        #telefone = isinstance(fatia , int)
        for b, fatias in enumerate(fatia):
            #print(i, valor)
            if fatias == "(11)":
                
                #ddd = (b+1)
                telefone = (b+1)
 

                whatsapp = fatia[b] + fatia[telefone]

                print(f' WhatsaApp da loja  é {whatsapp} ' )
                
                print("!!!!!" * 10)


                lista_telefone.append(whatsapp)    
        
    time.sleep(5)
    print('Dados Coletados')

    resultado = (lista_empresa + lista_telefone)

    print(100 * "&")

    navegador.find_element(By.XPATH, '//*[@id="pnnext"]/span[1]').click();
    print(f'estamos na página {numero}')
    
        
print('TRABALHO EXECUTADO.')

print(lista_empresa)
print(lista_telefone)

def merge(list1, list2): 
      
    merged_list = list(zip(list1, list2))  
    return merged_list 
      
list1 = lista_empresa
list2 = lista_telefone 
malandro = merge(list1, list2)
print(malandro)


dataframe = pd.DataFrame(malandro)
print(dataframe)
#dataframe.to_excel("res.xls")NÃO FUNCIONA .XLS
dataframe.to_excel("TesteItaimPaulista.xlsx" , index=False)

#site:www.instagram.com + "Itaim Paulista" + "WhatsApp" + "(11)"
```
